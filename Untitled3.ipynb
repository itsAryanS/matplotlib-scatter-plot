# Scatter Plot

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
num=[2,3,4,5,6,4,1]
plt.scatter(day,num, color="red")
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("Num")
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
num=[2,3,4,5,6,4,1]
sizes=[100,60,80,60,90,70,80]
colors=["y","b","r","g","g","g","g"]
plt.scatter(day,num,c=colors, s=sizes, alpha=0.9, marker="*", edgecolor="black", linewidth=0.8)
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("Num")
plt.show()
```

```python
import matplotlib.pyplot as plt
day=[1,2,3,4,5,6,7]
num=[2,3,4,5,6,4,1]
num2=[3,5,6,3,7,9,9]
sizes=[100,60,80,60,90,70,80]
colors=[10,29,30,29,56,20,30]
plt.scatter(day,num,c=colors, s=sizes,cmap="viridis")
plt.scatter(day,num2,color="r", s=sizes)
plt.colorbar()
t=plt.colorbar()
plt.title("Scatter Plot",fontsize=20)
plt.xlabel("Day")
plt.ylabel("Num")
plt.show()
```
